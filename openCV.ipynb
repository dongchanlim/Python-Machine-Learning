```
import cv2
from google.colab import files
import matplotlib.pyplot as plt

file = files.upload()
```

```
# image(practice.jpg) is changed into numpy array
from google.colab.patches import cv2_imshow

image = cv2.imread("practice2.jpg", flags = cv2.IMREAD_COLOR)
cv2_imshow(image)
```

```
plt_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(plt_image)
```

```
image.shape, plt_image.shape
```

```
glay_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(glay_image)
```

```
# interporation (resize)
#resize_image = cv2.resize(image, dsize = None, fx = 0.25, fy = 0.25)
resize_image = cv2.resize(image, dsize = (1000,800), interpolation = cv2.INTER_CUBIC)
cv2_imshow(resize_image)
```

```
# rotate the image
import numpy as np

width, height = resize_image.shape[:2]
M = cv2.getRotationMatrix2D(center = (width/2 , height/2), angle = -17, scale = 1.5)
M2 = np.float32([[1, 0, - 80],[0, 1, 30]])
image_trans = cv2.warpAffine(resize_image, M = M, dsize = (width , height))
image_trans = cv2.warpAffine(image_trans, M = M2, dsize = (width , height))

cv2_imshow(image_trans)
```

```
image_trans_grey = cv2.cvtColor(image_trans, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_trans_grey)
```

```
ret, thresh = cv2.threshold(image_trans_grey, 167, 255, 0)
thresh = cv2.cvtColor(thresh, cv2.COLOR_GRAY2BGR)
plt.imshow(thresh)
```
